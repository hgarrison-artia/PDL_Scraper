```python
import pdfplumber
import pandas as pd
```

```python
file_path = 'MT.pdf'

word_dict = []

with pdfplumber.open(file_path) as pdf:
    all_tables = []

    page_num = 0

    for page in pdf.pages:

        tables = page.extract_tables()

        for table in tables:
            df = pd.DataFrame(table)
            all_tables.append(df)

        words = page.extract_words()

        for word in words:
            word_dict.append({'Word': word['text'],
                             'x': word['x0'],
                             'Page': page_num})
        page_num += 1
```
