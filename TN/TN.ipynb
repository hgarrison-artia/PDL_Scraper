```python
import pandas as pd
import docx
```

```python
doc_path = 'TN.docx'

doc = docx.Document(doc_path)
data = []
for table in doc.tables:
    for row in table.rows:
        data.append([cell.text for cell in row.cells])

# Convert to DataFrame      
df = pd.DataFrame(data)
```
