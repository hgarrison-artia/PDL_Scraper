```python
import docx
import pandas as pd

doc_path = 'IA.docx'
doc = docx.Document(doc_path)

for idx, table in enumerate(doc.tables):
    print(f"Table {idx}:")
    data = []
    for row in table.rows:
        data.append([cell.text for cell in row.cells])
    df = pd.DataFrame(data)
    print(df)
    print("\n" + "="*40 + "\n")
```
